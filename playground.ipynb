```python
from manim import *

class demo(Scene):
    def construct(self):
        circle = Circle(radius=0.5, stroke_width=10, color=RED, fill_opacity=0.3)
        rect = SurroundingRectangle(circle, color=BLUE, corner_radius=0.1)
        text = Text("Manim")

        # self.add(NumberPlane())

        text.next_to(rect, UP, buff=0.5)

        self.play(Write(circle), Write(rect), Write(text))

        circ_and_rect = VGroup(circle, rect)

        self.play(text.animate.move_to([-4,0,0]))
        self.play(circ_and_rect.animate.move_to([4,0,0]))

        arrow = always_redraw(lambda: Line(start=circ_and_rect.get_left(), end=text.get_right(), buff=0.3).add_tip(tip_shape=StealthTip).add_tip(tip_shape=StealthTip, at_start=True))

        self.play(Write(arrow))

        self.play(Indicate(text, 1.5, color=ORANGE))

        self.play(Rotate(rect, angle=PI/2), ScaleInPlace(circle, scale_factor=2))

        self.play(circ_and_rect.animate.move_to(ORIGIN))

        self.play(FadeOut(arrow, text))

        self.play(ScaleInPlace(circle, 10), run_time=2)
        self.play(ScaleInPlace(rect, scale_factor=0))

        self.wait(1)


%manim -qm -v WARNING demo
```

```python
from manim import *

class CodeFromString(Scene):
    def construct(self):
        code = '''
            from manim import Scene, Square

            class FadeInSquare(Scene):
                def construct(self):
                    s = Square()
                    self.play(FadeIn(s))
                    self.play(s.animate.scale(2))
                    self.wait()
        '''

        rendered_code = Code(
            code_string=code,
            language="python",
            background="window",
            background_config={"stroke_color": "maroon"},
        )
        self.add(rendered_code)

%manim -qm -v WARNING CodeFromString
```

```python
from manim import *

class demo(Scene):
    def construct(self):
        plane = NumberPlane()
        self.add(plane)
        axes = Axes(x_range=(-12,12), y_range=(-4,4), x_length=12, y_length=4, tips=False, axis_config=axis_config).add_coordinates()

        x = axes.get_x_axis_label("x")
        y = axes.get_y_axis_label("y")
        
        self.play(Write(axes), Write(x), Write(y))
        
        coef = ValueTracker(1)
        coef2 = ValueTracker(1)

        curve = always_redraw(lambda: axes.plot(lambda x: np.cos(coef2.get_value()*x)*coef.get_value(), color=RED))        
        coef_text = always_redraw(lambda: MathTex(
            f"{coef.get_value():.1f} \\cos({coef2.get_value():.1f} x)", 
            font_size=40
        ).move_to(axes.c2p(6, 3)))   
             
        self.play(Write(coef_text))
        self.play(Write(curve))
        
        self.play(coef.animate.set_value(0.5), run_time=1)
        self.play(coef.animate.set_value(2), run_time=1)
        self.play(coef.animate.set_value(1), run_time=1)
        
        self.play(coef2.animate.set_value(0.5), run_time=1)
        self.play(coef2.animate.set_value(2), run_time=1)
        self.play(coef2.animate.set_value(1), run_time=1)


        self.wait(2)
        
%manim -qm -v WARNING demo
```
